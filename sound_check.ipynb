```python
import matplotlib.pyplot as plt # plotting
from scipy import signal # window signal
import matplotlib # plotting parameters
import pyanito.pyanito as pyano
matplotlib.rcParams.update({'font.size': 8}) # enlarge font size
```

# Sound check

Let's create a sinusoid wave and play it

```python
tf = 1 # note length
fs = 44100 # sampling rate
t = pyano.time_arr(tf, fs)
f0 = 100 # fundamental frequency (F0)
n = 1#5
y = pyano.harmonic(f0, t, n)
pyano.play(y, fs=fs)
```

Now let's visualise the wave and the power spectral density

```python
fig, ax = plt.subplots(2,1, figsize=(8, 4))
ax[0].plot(t, y)
ax[0].set_xlim(0.2, 0.3)
ax[0].set_xlabel("Time (s)")
ax[0].title.set_text("Waveform of the superposition of sinusoid "\
                      "waves with $F_0$ = {} Hz and {} harmonics".format(f0, n))

freqs, Pxx = signal.periodogram(y, fs=fs)
ax[1].plot(freqs, Pxx)
ax[1].set_xlim(0, f0*(n+1))
ax[1].set_xlabel("Frequency (Hz)")
ax[1].title.set_text("Power specral density")

plt.tight_layout()
```
